```python
fig, ax = plt.subplots()
plt.rc('ytick', labelsize=7)
plt.rc('xtick', labelsize=7)
fig.suptitle('Venus Bay Surface Sample Locations', y=0.775)

plt.subplot(121)
ss = 25

plt.scatter(lon,lat,s=ss)
plt.plot(lon,lat)
plt.xlabel('Longitude ($^\\circ$)')
plt.ylabel('Latitude ($^\\circ$)')

plt.axis('square')
plt.xticks([145.7906, 145.7916, 145.7926, 145.7936],
           ['145.7906', '145.7916', '145.7926','145.7936'])

plt.subplot(122)
plt.plot(dist,eh)
plt.scatter(dist,eh,s=ss)
plt.xlabel('Distance (m)')
plt.ylabel('Ellipsoidal Height (cm)')
plt.axis('square')
fig.tight_layout()
```

```python
ss = 20
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)

sls = plt.scatter(fcl,el,s=ss, label='Lutregala')
sws = plt.scatter(fcw,ew,s=ss, label='Wapengo')

plt.ylim(0,1.05)
plt.xlim(0,850)

plt.title('Foram Count at Elevation',fontsize=15)
plt.ylabel('Elevation (m AHD)',fontsize=12)
plt.xlabel('Count',fontsize=12)
plt.legend(frameon=False)
```

```python
forams = [fc, fcl, fcw]
fig = plt.figure(figsize =(5,7))
ax = fig.add_axes([0, 0, 1, 1])
fbp = ax.boxplot(forams)
plt.rc('ytick', labelsize=12)
plt.ylabel('Count',fontsize=20)
plt.title("Foram Count",fontsize=25)
ax.set_xticklabels(['Regional', 'Lutregala', 'Wapengo'],fontsize=20)
plt.show
```

```python
forams = [f1,f2,f4,f6,f7,f9]
plt.rc('ytick', labelsize=25)
fig = plt.figure(figsize =(10,7))
ax = fig.add_axes([0, 0, 1, 1])
fbp = ax.boxplot(forams)
plt.yscale('log')
plt.ylabel('%',fontsize=25)
plt.title("Relative Abundance of Common Foraminifera",fontsize=25)
ax.set_xticklabels(['T.salsa', 'H.wilberti', 'T.irregularis', 'E.macrescens', 'T.inflata', 'M.fusca'],fontsize=17)
plt.show
```
